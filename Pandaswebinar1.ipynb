```
#filter using boolean condition
```

```
bool_cond = newindex_df.index.str.startswith('A')
second_cond = newindex_df.turnout >=67
newindex_df[bool_cond & second_cond] # it filters out all the rows which don't match the output of bool
#newindex_df[bool_cond | second_cond] 
```

```
cond_obama = newindex_df.winner == 'Obama'
newindex_df[cond_obama]
```

```
cond_obama = newindex_df.winner == 'Romney'
newindex_df[~cond_obama] #negating 
```

```
# query method
```

```
newindex_df.query('turnout>70')
```

```
cond = 70
newindex_df.query('turnout > @cond') #@ tells panda dat it is a global variable
```

```
cond = 70
startwithvar = 'A'
newindex_df.query('turnout > @cond & index.str.startswith(@startwithvar)') # this will fail as query is based on a num 
```

```
cond = 70
startwithvar = 'A'
newindex_df.query('turnout > @cond & index.str.startswith(@startwithvar)',engine = 'python') # this will results in loss of efficiency as the eval logic for query is more efficient
```
